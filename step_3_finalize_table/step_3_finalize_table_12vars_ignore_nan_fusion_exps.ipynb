```python
import csv
import pandas as pd
import json
import os
import os.path as pathlib
import numpy as np
from copy import deepcopy
from glob import glob
from finalize_table_common_fun import table_one_hot_encoding

def do_conversion(dataset_type='random'):
    toproot_dir = '/ssd/kangdang/mmc_project_2023_submission'
    root_dir = pathlib.join(toproot_dir, 'main_exp_code', 'step1_preprocessing_redo')
    output_dir = pathlib.join(root_dir, 'preprocessed_data', dataset_type + '_processed_final_csv')
    os.makedirs(output_dir, exist_ok=True)
    img_only_csv_list = [pathlib.join(root_dir, 'preprocessed_data', 'train_table_orig_'+ dataset_type +'.csv'), 
                         pathlib.join(root_dir, 'preprocessed_data', 'val_table_orig_'+ dataset_type +'.csv'),
                         pathlib.join(root_dir, 'preprocessed_data', 'test_table_orig_'+ dataset_type + '.csv')]
    for img_level_csv in img_only_csv_list:
        patient_level_csv = pd.read_csv(pathlib.join(output_dir, pathlib.basename(img_level_csv).replace('.csv', '_12_vars_with_nan_intermediate_table_patient_level.csv')))
        patient_params = patient_level_csv[['param1', 'param2', 'param3', 'param4', 'param5', 'param6', 'param7', 'param8', 'param9', 'param10', 'param11', 'param12']]
        param_nan_counts = np.sum(pd.isna(patient_params), axis=1)
        patient_level_csv_no_nan = patient_level_csv[param_nan_counts == 0]
        print(len(patient_level_csv))
        print(len(patient_level_csv_no_nan))
        patient_level_csv_no_nan = table_one_hot_encoding(patient_level_csv_no_nan)
        print(pathlib.join(output_dir, pathlib.basename(img_level_csv).replace('.csv', '_12_vars_fusion_remove_any_nan_sample_patient_level.csv')))
        patient_level_csv_no_nan.to_csv(pathlib.join(output_dir, pathlib.basename(img_level_csv).replace('.csv', '_12_vars_fusion_remove_any_nan_sample_patient_level.csv')), index=False)
```

```python
do_conversion(dataset_type='random')
```
