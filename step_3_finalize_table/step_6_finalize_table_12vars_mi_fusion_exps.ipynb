```python
import csv
import pandas as pd
import json
import os
import os.path as pathlib
import numpy as np
from copy import deepcopy
from glob import glob
from finalize_table_common_fun import merge_imputation_data

def merge_imputation_data_main(orig_table_filename, imputation_data_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    orig_table = pd.read_csv(orig_table_filename)
    imputation_table_idx_list = (np.arange(20) + 1).astype(np.int64)
    for imputation_table_idx in imputation_table_idx_list:
        print(imputation_table_idx, len(imputation_table_idx_list))
        merged_table = merge_imputation_data(orig_table, imputation_data_dir, imputation_table_idx)
        merged_table_filename = orig_table_filename.replace('with_nan_intermediate_table', 'fusion_mi_full_idx_' + str(imputation_table_idx))
        merged_table_filename = pathlib.join(output_dir, pathlib.basename(merged_table_filename))
        merged_table.to_csv(merged_table_filename, index=False)

def do_conversion(dataset_type='random'):
    toproot_dir = '/ssd/kangdang/mmc_project_2023_submission'
    root_dir = pathlib.join(toproot_dir, 'main_exp_code', 'step1_preprocessing_redo')
    orig_table_dir = pathlib.join(root_dir, 'preprocessed_data', dataset_type + '_processed_final_csv')
    imputation_data_dir = pathlib.join(root_dir, 'preprocessed_data', 'mi_target')
    output_dir = pathlib.join(root_dir, 'preprocessed_data', dataset_type + '_processed_final_csv', 'fusion_mi')
    os.makedirs(output_dir, exist_ok=True)
    train_orig_name = pathlib.join(orig_table_dir, 'train_table_orig_' + str(dataset_type) + '_12_vars_with_nan_intermediate_table_patient_level.csv')
    val_orig_name = pathlib.join(orig_table_dir, 'val_table_orig_' + str(dataset_type) + '_12_vars_with_nan_intermediate_table_patient_level.csv')
    test_orig_name = pathlib.join(orig_table_dir, 'test_table_orig_' + str(dataset_type) + '_12_vars_with_nan_intermediate_table_patient_level.csv')
    print('train dataset')
    merge_imputation_data_main(train_orig_name, imputation_data_dir, output_dir)
    print('val dataset')
    merge_imputation_data_main(val_orig_name, imputation_data_dir, output_dir)
    print('test dataset')
    merge_imputation_data_main(test_orig_name, imputation_data_dir, output_dir)
```

```python
do_conversion(dataset_type='random')
```
