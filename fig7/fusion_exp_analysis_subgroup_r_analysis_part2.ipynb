```r
#### library(nsROC)
library(pROC)
library(caret)
library("PredictABEL")

auc_test <- function(final_fusion_result_csv, final_fusion_result_csv1){
    ori <- read.csv(final_fusion_result_csv, fileEncoding="gb18030")
    ori1 <- read.csv(final_fusion_result_csv1, fileEncoding="gb18030") 
    ren <- data.frame(ori[, 1:2])
    ren1 <- data.frame(ori1[, 1:2])
    names(ren) <- c("response1","p1")
    names(ren1) <- c("response2","p2")
    auc1 <- roc(response1~p1,data=ren,smooth=F)
    auc2 <- roc(response2~p2,data=ren1,smooth=F)
    print(roc.test(auc1,auc2, paired=FALSE))

    plot(auc1,print.auc=TRUE,print.thres=TRUE,main="多组ROC曲线比较",col="blue",identity.col="blue",identity.lty=1,identity.lwd=1)
    plot(auc2,print.auc=TRUE,print.thres=TRUE,add=TRUE,col="red")
    }
```

```r
output_dir = '/ssd/kangdang/mmc_project_2023_submission/main_exp_code/step4_exp_analysis/fig7/data/'
save_name1 = 'subgroup_female.csv'
save_name2 = 'subgroup_male.csv'
auc_test(paste(output_dir, save_name1, sep=""), paste(output_dir, save_name2, sep=""))
```
