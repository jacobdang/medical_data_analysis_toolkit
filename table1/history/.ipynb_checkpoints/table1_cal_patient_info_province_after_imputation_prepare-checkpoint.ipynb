```python
# obtain average statistical values
import pandas as pd
import numpy as np
import os.path as pathlib
from glob import glob
from scipy import stats as st
```

```python
root_dir = '/ssd/kangdang/mmc_project_final_v6/main_exp_code/step1_preprocessing_redo'
random_csv_list = [pathlib.join(root_dir, 'preprocessed_data', 'random_processed_final_csv', 'fusion_mi', 'train_table_orig_random_12_vars_fusion_mi_full_idx_*.csv'), 
                   pathlib.join(root_dir, 'preprocessed_data', 'random_processed_final_csv', 'fusion_mi', 'val_table_orig_random_12_vars_fusion_mi_full_idx_*.csv'),
                   pathlib.join(root_dir, 'preprocessed_data', 'random_processed_final_csv', 'fusion_mi', 'test_table_orig_random_12_vars_fusion_mi_full_idx_*.csv')]
```

```python
def process_table_main(table_all):
    all_table_numerical = list()
    all_table_categorical = list()

    params = [ 'param1', 'param2', 'param3', 'param4',
           'param5', 'param6', 'param7', 'param8', 'param9', 'param10', 'param35',
           'param11_1', 'param11_2', 'param11_3', 'param11_4', 'param11_5',
           'param11_6', 'param11_7', 'param12_1', 'param12_2', 'param12_3',
           'param12_4', 'param12_5']
    
    numerical_params = ['param2', 'param3', 'param4', 'param5', 'param6', 'param7',  'param35']
    categorical_params =  ['param1', 'param8', 'param9', 'param10', 'param11', 'param12']
    new_table_non_params = ["patient_id", "left_img_id", "right_img_id", "left_img_quality", "right_img_quality", "province"]

    for curr_table in table_all:
        curr_table_params = pd.read_csv(curr_table )[params]
        curr_table_params['param1'] = curr_table_params['param1'] + 1
        curr_table_params_param11 = curr_table_params[['param11_1', 'param11_2', 'param11_3', 'param11_4', 'param11_5', 'param11_6', 'param11_7']]
        curr_table_params_param11_renamed = curr_table_params_param11.rename(columns={'param11_1':1, 'param11_2':2, 'param11_3':3, 'param11_4':4,  'param11_5':5,  'param11_6':6,  'param11_7':7})
        curr_table_params_param11 = pd.from_dummies(curr_table_params_param11_renamed)
        curr_table_params['param11'] = curr_table_params_param11 
        curr_table_params_param12 = curr_table_params[['param12_1', 'param12_2', 'param12_3', 'param12_4', 'param12_5']]
        curr_table_params_param12_renamed = curr_table_params_param12.rename(columns={'param12_1':1, 'param12_2':2, 'param12_3':3, 'param12_4':4, 'param12_5':5})
        curr_table_params_param12 = pd.from_dummies(curr_table_params_param12_renamed)
        curr_table_params['param12'] = curr_table_params_param12
        curr_table_numerical = curr_table_params[numerical_params].to_numpy()
        curr_table_categorical = curr_table_params[categorical_params].to_numpy()
        all_table_numerical.append(curr_table_numerical)
        all_table_categorical.append(curr_table_categorical)
        
    all_table_numerical = np.array(all_table_numerical)
    all_table_categorical = np.array(all_table_categorical)
    all_table_numerical_mean = np.mean(all_table_numerical, axis=0)
    all_table_categorical_mode = st.mode(all_table_categorical, axis=0)[0][0]
    new_table = pd.read_csv(table_all[0])
    new_table_non_params = new_table[new_table_non_params]
    new_table_numerical_params = pd.DataFrame(all_table_numerical_mean, columns=numerical_params)
    new_table_categorical_params = pd.DataFrame(all_table_categorical_mode, columns=categorical_params)
    new_table_merge = pd.concat((new_table_non_params, new_table_numerical_params,new_table_categorical_params ), axis=1)
    return new_table_merge
```

```python
new_train_table = process_table_main(glob(random_csv_list[0]))
new_val_table = process_table_main(glob(random_csv_list[1]))
new_test_table = process_table_main(glob(random_csv_list[2]))
```

```python
# std/mean on 47269 patients
# param 1 gender, male 27110, female 20159
# param 2 age 54.29±11.60
# param 3 舒张压 diastolic blood pressure, 77.34±11.34
# param 4 收缩压 Systolic blood pressure, 131.40±19.06
# param 5 heart rate 
# 80.82604745973013 -> recovered mean
# 11.84815122304001 -> recovered std

# param 6  BMI, 25.62±3.77
# param 7 病程（月）Duration of diabetes, years 6.80±6.71
# param 8 您是否有明确诊断的高血压？ History of hypertension 16 266 (39.8)
# param 9 您是否患有高血脂？ History of dyslipidemia 11 361 (27.9)
# param 10 您是否患有心血管疾病？ 
# param 11 吸烟分类
# param 12 饮酒分类 [non-drinker,former drinker(light or heavy),current drinker(light or heavy)]
# HbA1c,% 9.06±2.27

random_train_csv = pd.read_csv(random_csv_list[0])
print(random_train_csv['param10'].value_counts())
print(random_train_csv['param11'].value_counts())
print(random_train_csv['param12'].value_counts())

param_transfer_dict = dict()
param_transfer_dict['param2'] = [54.29, 11.60]
param_transfer_dict['param3'] = [77.34, 11.34]
param_transfer_dict['param4'] = [131.40, 19.06]
param_transfer_dict['param5'] = [80.826, 11.848]
param_transfer_dict['param6'] = [25.62, 3.77]
param_transfer_dict['param7'] = [6.80, 6.71]
param_transfer_dict['param35'] = [1634.14, 349.12]
```

```python
def modify_helper2(array, th):
    array_pos = array >= th
    array_neg = array <th
    array[array_pos] = 1
    array[array_neg] = 0
    return array

def get_dataset_stats(csv_file_list):
    csv_list = list()
    for csv_file in csv_file_list:
        curr_csv = pd.read_csv(csv_file)
        csv_list.append(curr_csv)
    curr_csv = pd.concat(csv_list)

    for idx in list(range(12)) + [34]:
        if idx == 10: # smoking
            curr_csv['param11'] = modify_helper2(curr_csv['param11'], 2)
        if idx == 11: # drinking
            curr_csv['param12'] = modify_helper2(curr_csv['param12'], 2)
        param_str = 'param' + str(idx + 1)
        print(param_str)
        if param_str not in param_transfer_dict.keys():
            print(curr_csv[param_str].value_counts())
            print(curr_csv[param_str].value_counts(normalize=True))
            missing_value_no = int(np.sum(curr_csv[param_str].isnull()))
            print('missing: ' + str(missing_value_no) + ', perc=' + str(round(float(missing_value_no/len(curr_csv)),3)))
        else:
            curr_param_values = np.array(list(curr_csv[param_str]))
            curr_mean, curr_std = param_transfer_dict[param_str]
            curr_param_values = curr_param_values * curr_std + curr_mean
            print('mean is : ' + str(round(np.nanmean(curr_param_values), 2)) + ', std is : ' + str(round(np.nanstd(curr_param_values), 2)))
            missing_value_no = int(np.sum(curr_csv[param_str].isnull()))
            print('missing: ' + str(missing_value_no) + ', perc=' + str(round(float(missing_value_no/len(curr_csv)),3)))
```

```python
# random train + val
get_dataset_stats([random_csv_list[0], random_csv_list[1]])
```

```python
# random test
get_dataset_stats([random_csv_list[2]])
```
