```python
def get_dataset_stats(csv_file_list):
    csv_list = list()
    for csv_file in csv_file_list:
        curr_csv = pd.read_csv(csv_file)
        csv_list.append(curr_csv)
    curr_csv = pd.concat(csv_list)
    for idx in list(range(12)) + [34]:
        if idx == 10: # smoking
            curr_csv['param11'] = modify_helper2(curr_csv['param11'], 2)
        if idx == 11: # drinking
            curr_csv['param12'] = modify_helper2(curr_csv['param12'], 2)
        param_str = 'param' + str(idx + 1)
        print(param_str)
        if param_str not in param_transfer_dict.keys():
            print(curr_csv[param_str].value_counts())
            print(curr_csv[param_str].value_counts(normalize=True))
            missing_value_no = int(np.sum(curr_csv[param_str].isnull()))
            print('missing: ' + str(missing_value_no) + ', perc=' + str(round(float(missing_value_no/len(curr_csv)),2)))
        else:
            curr_param_values = np.array(list(curr_csv[param_str]))
            curr_mean, curr_std = param_transfer_dict[param_str]
            curr_param_values = curr_param_values * curr_std + curr_mean
            print('mean is : ' + str(round(np.nanmean(curr_param_values), 2)) + ', std is : ' + str(round(np.nanstd(curr_param_values), 2)))
            missing_value_no = int(np.sum(curr_csv[param_str].isnull()))
            print('missing: ' + str(missing_value_no) + ', perc=' + str(round(float(missing_value_no/len(curr_csv)),2)))


def modify_helper2(array, th):
    array_pos = array >= th
    array_neg = array <th
    array[array_pos] = 1
    array[array_neg] = 0
    return array
            
def get_dataset_details(csv_file_list, param_idx, ignore_nan=True):
    csv_list = list()
    for csv_file in csv_file_list:
        curr_csv = pd.read_csv(csv_file)
        csv_list.append(curr_csv)
    curr_csv = pd.concat(csv_list)
    
    param_str = 'param35'
    curr_param_values = np.array(list(curr_csv[param_str]))
    curr_mean, curr_std = param_transfer_dict[param_str]
    curr_param_values = curr_param_values * curr_std + curr_mean
    param_str = 'param99'
    curr_csv[param_str] = curr_param_values>=1400
    
    param_str = 'param' + str(param_idx + 1)
    print(param_str)
    if param_str not in param_transfer_dict.keys():
        if ignore_nan:
            isnull = curr_csv[param_str].isnull()
            result_data = curr_csv[param_str][~isnull]
        return ["categorical", list(result_data)]
    else:
        if ignore_nan:
            isnull = curr_csv[param_str].isnull()
            result_data = curr_csv[param_str][~isnull]
        curr_param_values = np.array(list(result_data))
        curr_mean, curr_std = param_transfer_dict[param_str]
        curr_param_values = curr_param_values * curr_std + curr_mean
        return ["numerical", list(curr_param_values)]

# random train + val
get_dataset_stats([random_csv_list[0], random_csv_list[1]])
```

```python
# random train + val
# std/mean on 47269 patients
# param 1 gender, male 27110, female 20159
# param 2 age 54.29±11.60
# param 3 舒张压 diastolic blood pressure, 77.34±11.34
# param 4 收缩压 Systolic blood pressure, 131.40±19.06
# param 5 heart rate 
# 80.82604745973013 -> recovered mean
# 11.84815122304001 -> recovered std

# param 6  BMI, 25.62±3.77
# param 7 病程（月）Duration of diabetes, years 6.80±6.71
# param 8 您是否有明确诊断的高血压？ History of hypertension 16 266 (39.8)
# param 9 您是否患有高血脂？ History of dyslipidemia 11 361 (27.9)
# param 10 您是否患有心血管疾病？ 
# param 11 吸烟分类
# param 12 饮酒分类 [non-drinker,former drinker(light or heavy),current drinker(light or heavy)]
# HbA1c,% 9.06±2.27

import scipy

def get_contingency_table(tr_val_value, te_value):
    tr_val_dict = Counter(tr_val_value)
    te_dict = Counter(te_value)
    contingency_table = list()
    tr_val_dict_keys_sorted = list(np.sort(list(tr_val_dict.keys())))
    for tr_key_idx in tr_val_dict_keys_sorted:
        tr_value = tr_val_dict[tr_key_idx]
        te_value = te_dict[tr_key_idx]
        contingency_table.append((tr_value, te_value))
    return contingency_table
        
from collections import Counter
def get_stat_test():
    for i in list(range(12)) + [34] + [98]:
        tr_val = get_dataset_details([random_csv_list[0], random_csv_list[1]], i)
        te = get_dataset_details([random_csv_list[2]], i)
        if i == 10: # smoking
            tr_val[1] = modify_helper2(np.array(tr_val[1]), 2)
            te[1] = modify_helper2(np.array(te[1]),  2)
        if i == 11: # drinking
            tr_val[1] = modify_helper2(np.array(tr_val[1]), 2)
            te[1] = modify_helper2(np.array(te[1]), 2)    
        if tr_val[0] == "numerical":
            norm_test_result = scipy.stats.kstest(tr_val[1], cdf="norm")
            if norm_test_result.pvalue > 0.05:
                print('normal distribution')
                result = scipy.stats.ttest_ind(tr_val[1], te[1])
            else:
                print('not normal distribution')
                result = scipy.stats.ranksums(tr_val[1], te[1])
            print(result)
        if tr_val[0] == "categorical":
            contingency_table = get_contingency_table(tr_val[1], te[1])
            print(contingency_table)
            result = scipy.stats.chi2_contingency(contingency_table)
            print(result)
get_stat_test()
```
