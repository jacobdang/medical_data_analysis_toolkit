```python
def get_statistic_info(test_gt_list, test_pred_list):
    stats_info = cal_ci95(test_gt_list, test_pred_list, 0.8)
    auc95 = stats_info[0]
    fpr, tpr, threshold = metrics.roc_curve(test_gt_list, test_pred_list)
    auc = metrics.auc(fpr, tpr)
    return auc95, fpr, tpr, auc

def get_info(csv_file, idx):
    test_pred_set = csv_file['param' + str(idx)]
    test_gt_set = csv_file['gt']
    auc95, fpr, tpr, auc = get_statistic_info(test_gt_set, test_pred_set)
    return auc95, auc
```

```python
metric_auc_list = list()
metric_auc_95_list = list()
fusion_auc_list = list()
fusion_auc_95_list = list()
indicator_name_list = ['Gender', 'Age', 'Diastolic Bp', 'Systolic Bp',  'Heart Rate', 'BMI', 
                       'Course of Diabetes', 'Hypertension', 
                       'Hyperlipidemia', 'Cardiovascular Disease',
                       'Classification of smoker', 'Classification of drinker', 'Community clinic scenario']
selected_idx = [1, 6, 3, 7, 2]
indicator_name_list = list(np.array(indicator_name_list)[np.array(selected_idx)]) + ['Community clinic scenario']

for i in selected_idx:
    auc95, auc = get_info(external_test_metric, i + 1)
    metric_auc_list.append(auc)
    metric_auc_95_list.append(auc95)
auc95, auc = get_info(external_test_metric_scenario, '_999')
metric_auc_list.append(auc)
metric_auc_95_list.append(auc95)


for i in selected_idx:
    auc95, auc = get_info(external_test_fusion, i + 1)
    fusion_auc_list.append(auc)
    fusion_auc_95_list.append(auc95)
auc95, auc = get_info(external_test_fusion_scenario, '_999')
fusion_auc_list.append(auc)
fusion_auc_95_list.append(auc95)
    
metric_auc_95 = np.array(metric_auc_95_list)
metric_auc_95[:,0] = np.array(metric_auc_list) - metric_auc_95[:,0]
metric_auc_95[:,1] = metric_auc_95[:, 1] - np.array(metric_auc_list)
metric_auc_95_new = metric_auc_95

fusion_auc_95 = np.array(fusion_auc_95_list)
fusion_auc_95[:,0] = np.array(fusion_auc_list) - fusion_auc_95[:,0]
fusion_auc_95[:,1] = fusion_auc_95[:,1] - np.array(fusion_auc_list)
fusion_auc_95_new = fusion_auc_95
```

```python
import matplotlib.pyplot as plt
def plot_fun(img_auc_list, img_auc_95_new, img_auc_95_list, 
             fusion_auc_list, fusion_auc_95_new, fusion_auc_95_list,
             indicator_name_list, title, 
             xlim=[0.675, 0.8], color1='lightsteelblue', color2='orange'):
    my_dpi=96
    fig, ax = plt.subplots(figsize=(480 * 5/my_dpi, 480/my_dpi), dpi=my_dpi)
    plt.style.use('seaborn-darkgrid')
    barWidth = 1
    
    bars1 = list(np.array(img_auc_list))
    yer1 = img_auc_95_new.transpose() 
    r1 = 5 * np.arange(len(img_auc_95_new)) 
    print(r1)
    barplt1 = plt.bar(r1, bars1, width = 2, color = color1, alpha=1, 
            edgecolor = 'black', yerr=yer1, capsize=3)
    label1 = ["{auc:.3f}\n[{aucl:.3f}, \n{aucr:.3f}]".
             format(auc=_, aucl=lr[0], aucr=lr[1]) 
             for (_, lr) in zip(np.array(img_auc_list), 
                                np.array(img_auc_95_list))]
    #label1 = ["{auc:.3f}".
    #         format(auc=_) for _ in np.array(img_auc_list)]   
    ax.bar_label(barplt1, label1, fontsize=18)
    
    bars2 = list(np.array(fusion_auc_list))
    yer2 = fusion_auc_95_new.transpose() 
    r2 = 5  * np.arange(len(fusion_auc_95_new)) + 2
    barplt2 = plt.bar(r2, bars2, width = 2, color = color2, alpha=1, 
            edgecolor = 'black', yerr=yer2, capsize=3)
    label2 = ["{auc:.3f}\n[{aucl:.3f}, \n{aucr:.3f}]".
             format(auc=_, aucl=lr[0], aucr=lr[1]) 
             for (_, lr) in zip(np.array(fusion_auc_list), 
                                np.array(fusion_auc_95_list))]
    #label2 = ["{auc:.3f}".
    #         format(auc=_) for _ in np.array(fusion_auc_list)]       
    ax.bar_label(barplt2, label2, fontsize=18)

    # general layout
    plt.xticks([5 * r + 1  for r in range(len(bars1))], indicator_name_list, rotation = 15, size = 20)
    plt.yticks(fontsize=26)
    plt.ylabel('auc')
    plt.ylim(xlim)
    if title:
        plt.title(title, loc='left', fontsize=32, fontweight=0, color='black')
    ax=plt.gca()
    ax.xaxis.label.set_fontsize(36)
    ax.yaxis.label.set_fontsize(36)
    ax.legend([barplt1, barplt2], ['Factor Model', 'Fusion Model'], ncol=2,  fontsize= 18, loc=[0,1])
    plt.show() 
```

```python
plot_fun(metric_auc_list[0:6], metric_auc_95_new[0:6], metric_auc_95_list[0:6], 
         fusion_auc_list[0:6], fusion_auc_95_new[0:6], fusion_auc_95_list[0:6],
         indicator_name_list[0:6], title=None, 
         xlim=[0.3, 1.0])
```
